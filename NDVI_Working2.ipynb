```python
import os
base_path= r'C:\Users\jeidmann\Documents\GIS_Files\Complete2016Landsat8'
directory=os.listdir(base_path)
```

```python
#Basic Mosaic Inputs
#mdname = r"C:\Users\jeidmann\Documents\GIS_Files\PRDataExplore.gdb\LandsatFeb2016"
rastype = "Raster Dataset"
#inpath = list2[0], list2[1]
updatecs = "UPDATE_CELL_SIZES"
updatebnd = "UPDATE_BOUNDARY"
updateovr = "UPDATE_OVERVIEWS"
maxlevel = "2"
maxcs = "#"
maxdim = "#"
spatialref = "#"
inputdatafilter = "*.tif"
subfolder = "NO_SUBFOLDERS"
duplicate = "EXCLUDE_DUPLICATES"
buildpy = "BUILD_PYRAMIDS"
calcstats = "CALCULATE_STATISTICS"
buildthumb = "NO_THUMBNAILS"
comments = "Add Raster Datasets"
forcesr = "#"

for mosaic in directory:
    temp_path=base_path+'\\'+mosaic
    temp_files=os.listdir(temp_path)
    list_b4=[]
    list_b5=[]
    for s in temp_files:
        if 'band4' in s:
            list_b4.append(temp_path+'\\'+s)
        if 'band5' in s:
            list_b5.append(temp_path+'\\'+s)
    #print(len(list_b4))
    #print(len(list_b5))
    if len(list_b4)==2:
        inpath=list_b4[0],list_b4[1]
        mdname=base_gdb+'\\'+mosaic+'_Band4'
        arcpy.AddRastersToMosaicDataset_management(
         mdname,  rastype, inpath, updatecs, updatebnd, updateovr,
         maxlevel, maxcs, maxdim, spatialref, inputdatafilter,
         subfolder, duplicate, buildpy, calcstats, 
         buildthumb, comments, forcesr)
    if len(list_b4)==3:
        inpath=list_b4[0],list_b4[1],list_b4[2]
        mdname=base_gdb+'\\'+mosaic+'_Band4'
        arcpy.AddRastersToMosaicDataset_management(
         mdname,  rastype, inpath, updatecs, updatebnd, updateovr,
         maxlevel, maxcs, maxdim, spatialref, inputdatafilter,
         subfolder, duplicate, buildpy, calcstats, 
         buildthumb, comments, forcesr)
    if len(list_b4)==4:
        inpath=list_b4[0],list_b4[1],list_b4[2],list_b4[3]
        mdname=base_gdb+'\\'+mosaic+'_Band4'
        arcpy.AddRastersToMosaicDataset_management(
         mdname,  rastype, inpath, updatecs, updatebnd, updateovr,
         maxlevel, maxcs, maxdim, spatialref, inputdatafilter,
         subfolder, duplicate, buildpy, calcstats, 
         buildthumb, comments, forcesr)
    if len(list_b5)==2:
        inpath=list_b5[0],list_b5[1]
        mdname=base_gdb+'\\'+mosaic+'_Band5'
        arcpy.AddRastersToMosaicDataset_management(
         mdname,  rastype, inpath, updatecs, updatebnd, updateovr,
         maxlevel, maxcs, maxdim, spatialref, inputdatafilter,
         subfolder, duplicate, buildpy, calcstats, 
         buildthumb, comments, forcesr)
    if len(list_b5)==3:
        inpath=list_b5[0],list_b5[1],list_b5[2]
        mdname=base_gdb+'\\'+mosaic+'_Band5'
        arcpy.AddRastersToMosaicDataset_management(
         mdname,  rastype, inpath, updatecs, updatebnd, updateovr,
         maxlevel, maxcs, maxdim, spatialref, inputdatafilter,
         subfolder, duplicate, buildpy, calcstats, 
         buildthumb, comments, forcesr)
    if len(list_b5)==4:
        inpath=list_b5[0],list_b5[1],list_b5[2],list_b5[3]
        mdname=base_gdb+'\\'+mosaic+'_Band5'
        arcpy.AddRastersToMosaicDataset_management(
         mdname,  rastype, inpath, updatecs, updatebnd, updateovr,
         maxlevel, maxcs, maxdim, spatialref, inputdatafilter,
         subfolder, duplicate, buildpy, calcstats, 
         buildthumb, comments, forcesr)
```

Calculate the NDVI values:

```python
import arcpy
arcpy.CheckOutExtension("Spatial")
from arcpy import env
from arcpy.sa import *

env.workspace = r'C:\Users\jeidmann\Documents\GIS_Files\Landsat2014_clipped.gdb'
base_path= r'C:\Users\jeidmann\Documents\GIS_Files\Landsat2014_clipped.gdb'
Month=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
Year='2014'
Month2=[]
for name in Month:
    Month2.append(name+Year)
    Month2.append(name+'_b'+Year)

a2018=(1,12,-1)
a2017=(1,-1)
a2016=(1,-1)
a2015=(1,8,-1)
a2014=(3,12,-1)

for inx in a2014:
    print(Month2.pop(inx))

for category in Month2:
    band4=category+"_Band4"+"revised"
    band5=category+"_Band5"+"revised"
    output_raster=category+"_NDVI"
    NDVI=(Raster(band5)-Raster(band4))/(Raster(band5)+Raster(band4))
    #NDVI=(band5-band4)/(band5+band4)
    NDVI.save(output_raster)
```
